# Data Preprocessing from a CSV file

```python
import pandas as pd
df = pd.read_csv('asset_monitoring_sample_dataset.csv')
df.head()
```

Visual Representation Tools:

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

# Apply Isolation Forest for Anomaly Detection

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# Step 1: Select only the numeric columns (you may skip datetime, string col+
# umns)
numeric_df = df.select_dtypes(include=[np.number])

# Step 2: Fit Isolation Forest
iso_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
df['anomaly'] = iso_forest.fit_predict(numeric_df)

# Step 3: Label anomalies
# -1 means anomaly, 1 means normal
df['anomaly'] = df['anomaly'].map({1: 'Normal', -1: 'Anomaly'})

# Step 4: Count how many anomalies were detected
print(df['anomaly'].value_counts())
```

# Visualize the Anomalies

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Take any 2 parameters (example)
x = 'temperature'
y = 'pressure'

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=x, y=y, hue='anomaly', palette={'Normal': 'blue', 'Anomaly': 'red'})
plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel(x)
plt.ylabel(y)
plt.grid(True)
plt.show()
```

# Credit Card Fraud Detection (Kaggle)

```python
import pandas as pd

cc_df = pd.read_csv('creditcard.csv')
cc_df.head()
```

```python
cc_df.describe()
```

```python
cc_df.info()
```

```python
cc_df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cc_df.hist(figsize=(16,12))
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# Step 1: Select only the numeric columns (you may skip datetime, string col+
# umns)
numeric_cc_df = cc_df.select_dtypes(include=[np.number])

# Step 2: Fit Isolation Forest 
# for n_estimators: Higher value generally improves performance but increases training time.(Here 250K rows are there)
iso_forest = IsolationForest(n_estimators=250, contamination=0.01, random_state=42)
cc_df['anomaly'] = iso_forest.fit_predict(numeric_cc_df)

# Step 3: Label anomalies
# -1 means anomaly, 1 means normal
cc_df['anomaly'] = cc_df['anomaly'].map({1: 'Normal', -1: 'Anomaly'})

# Step 4: Count how many anomalies were detected
print(cc_df['anomaly'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Take any 2 parameters (example)
x = 'V12'
y = 'V4'

plt.figure(figsize=(10, 6))
sns.scatterplot(data=cc_df, x=x, y=y, hue='anomaly', palette={'Normal': 'blue', 'Anomaly': 'red'})
plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel(x)
plt.ylabel(y)
plt.grid(True)
plt.show()
```
